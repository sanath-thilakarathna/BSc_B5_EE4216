# Linear Regression: Predicting House Prices

In this notebook, we introduce **Linear Regression**, one of the most fundamental machine learning algorithms.

We will start with a **simple problem**:

Predict the **price of a house** using **only one feature — the house area**.

This example demonstrates the basic concept of regression before extending to more complex models.

## 1. Import Required Libraries

We first import the Python libraries required for data handling, visualization, and machine learning.

- **pandas** – for data manipulation
- **numpy** – for numerical computations
- **matplotlib** – for visualization
- **scikit-learn** – for machine learning models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

## 2. Load the Dataset

For this example, we use a small dataset that contains:

- **House area (square meters)**
- **House price**

This dataset represents a simplified housing market where price depends mainly on the area of the house.

```python
df = pd.read_csv("house_price_01.csv")
df
```

## 3. Understanding the Dataset

Our dataset contains two variables:

| Variable | Description |
|--------|-------------|
| area | Size of the house in square meters |
| price | Selling price of the house |

The goal of the model is to learn the relationship between **house area** and **house price**.

```python
df.info()
```

## 4. Visualizing the Relationship

Before training a machine learning model, it is important to visualize the data.

A **scatter plot** helps us understand whether a relationship exists between the variables.

If the data points roughly follow a straight line, a **linear regression model** may work well.

```python
plt.scatter(df["area"], df["price"])

plt.xlabel("House Area (m²)")
plt.ylabel("House Price")
plt.title("House Price vs Area")

plt.show()
```

## 5. Linear Regression Model

Linear regression models the relationship between an input variable and an output variable using a straight line.

The equation of a simple linear regression model is:

$$
y = \beta_0 + \beta_1 x
$$

Where:

- $y$ = predicted house price  
- $x$ = house area  
- $\beta_0$ = intercept  
- $\beta_1$ = slope of the line

## 6. Preparing the Data

Machine learning models require input data in a specific format.

We define:

- **X** → input feature (house area)
- **y** → target variable (house price)

```python
X = df[["area"]]
y = df["price"]
```

## 7. Training the Linear Regression Model

We now train the linear regression model using the dataset.

The model learns the best line that fits the data by minimizing the prediction error.

```python
model = LinearRegression()

model.fit(X, y)
```

## 8. Model Parameters

After training, the model estimates two important parameters:

- **Intercept** (\(\beta_0\))
- **Slope** (\(\beta_1\))

These parameters define the regression equation.

```python
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

## 9. Visualizing the Regression Line

We can visualize the fitted regression line together with the original data points.

The red line represents the model's predicted relationship between house area and house price.

```python
plt.scatter(X, y)

plt.plot(X, model.predict(X), color="red")

plt.xlabel("House Area")
plt.ylabel("House Price")
plt.title("Linear Regression Fit")

plt.show()
```

## 10. Predicting House Price

Once the model is trained, it can be used to predict house prices for new house areas.

For example, we can estimate the price of a house with an area of **140 square meters**.

```python
new_house = np.array([[140]])

predicted_price = model.predict(new_house)

print("Predicted price:", predicted_price[0])
```

```python
new_house = pd.DataFrame({"area":[140]})

predicted_price = model.predict(new_house)

print("Predicted price:", predicted_price[0])
```
