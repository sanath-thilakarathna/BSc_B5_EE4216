# Encoding Methods in Machine Learning

Machine Learning algorithms cannot directly work with categorical (text) data.  
Therefore, categorical variables must be converted into **numerical representations** before training models.

This process is called **Encoding**.

In this notebook we will explore several encoding techniques commonly used in machine learning:

1. Label Encoding
2. One-Hot Encoding
3. Ordinal Encoding
4. Binary Encoding
5. Target Encoding

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

## Sample Dataset

We create a small dataset containing categorical variables such as:

- City
- Size
- Color

These will be used to demonstrate encoding methods.

```python
data = pd.DataFrame({
    "City": ["Colombo", "Kandy", "Galle", "Colombo", "Kandy"],
    "Size": ["Small", "Medium", "Large", "Medium", "Small"],
    "Color": ["Red", "Blue", "Green", "Blue", "Red"]
})

data
```

```python
label_encoder = LabelEncoder()

data["City_LabelEncoded"] = label_encoder.fit_transform(data["City"])

data
```

## One-Hot Encoding

One-Hot Encoding converts each category into **separate binary columns**.

Example:

| City | Colombo | Kandy | Galle |
|-----|-----|-----|-----|
| Colombo | 1 | 0 | 0 |

Advantages:
- No false ordering

Limitation:
- Can create **many columns** when categories are large.

```python
encoder = OneHotEncoder(sparse_output=False)

onehot = encoder.fit_transform(data[["City"]])

onehot = pd.DataFrame(
    onehot,
    columns=encoder.get_feature_names_out(["City"])
)

onehot
```

```python
data_onehot = pd.concat([data, onehot], axis=1)

data_onehot
```

## Ordinal Encoding

Ordinal Encoding is used when categories have **natural ordering**.

Example:

Small < Medium < Large

```python
size_order = [["Small", "Medium", "Large"]]

ordinal_encoder = OrdinalEncoder(categories=size_order)

data["Size_Ordinal"] = ordinal_encoder.fit_transform(data[["Size"]])

data
```

## Binary Encoding

Binary Encoding converts categories into integers and then represents them using **binary numbers**.

This method helps reduce the number of columns compared to One-Hot Encoding.

## Target Encoding

Target Encoding replaces each category with the **mean value of the target variable**.

This method is powerful but must be used carefully to avoid **data leakage**.

```python
data_target = pd.DataFrame({
    "City": ["Colombo", "Kandy", "Galle", "Colombo", "Kandy"],
    "Price": [200000, 150000, 180000, 220000, 160000]
})

mean_encoding = data_target.groupby("City")["Price"].mean()

data_target["City_TargetEncoded"] = data_target["City"].map(mean_encoding)

data_target
```
