```python
import cv2

print("OpenCV version:")
print(cv2.__version__)

print("✅ OpenCV is installed correctly")
```

```python
# Create a simple grayscale image (100x100)
test_img = np.zeros((100, 100), dtype=np.uint8)
test_img[25:75, 25:75] = 255

print("Image shape:", test_img.shape)
print("Data type:", test_img.dtype)
```

```python
plt.imshow(test_img, cmap='gray')
plt.title("Test Image Display")
plt.axis('off')
plt.show()

print("✅ Image display works")
```

```python
# Apply Gaussian blur
blurred = cv2.GaussianBlur(test_img, (5, 5), 0)

plt.figure(figsize=(6,3))

plt.subplot(1,2,1)
plt.imshow(test_img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred")
plt.axis('off')

plt.show()

print("✅ OpenCV image processing works")
```
