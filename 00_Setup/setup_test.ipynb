```python
import sys

print("Python version:")
print(sys.version)

major, minor = sys.version_info[:2]
if major == 3 and minor == 10:
    print("✅ Python version is correct (3.10)")
else:
    print("❌ WARNING: Python version is NOT 3.10")
```
